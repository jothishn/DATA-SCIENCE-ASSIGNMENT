# Q.1 Delivery_time -> Predict delivery time using sorting time

```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("C:\\Users\\JOTHISH N\\Desktop\\DS\\SIMPLE LINEAR REGRESSION\\delivery_time.csv")
```

```python
data.head()
```

# EDA and Data Visualization

```python
data.info()
```

```python
import seaborn as sns
sns.distplot(data['Sorting Time'])
```

```python
import seaborn as sns
sns.distplot(data['Delivery Time'])
```

# Feature Engineering

```python
# Renaming Columns
df = data.rename({'Delivery Time':'delivery_time', 'Sorting Time':'sorting_time'}, axis=1)
df
```

# Correlation

```python
data.corr()
```

```python
sns.regplot(x="sorting_time", y="delivery_time", data=df);
```

# Model Building

```python
import statsmodels.formula.api as smf
model = smf.ols("delivery_time~sorting_time",data = df).fit()
```

# Model Testing

```python
#Coefficients
model.params
```

```python
#t and p-Values
print(model.tvalues, '\n', model.pvalues)    
```

```python
#R squared values
(model.rsquared,model.rsquared_adj)
```

# Model Predictions

```python
# Manual prediction for say sorting time 5
delivery_time = (6.582734) + (1.649020)*(5)
delivery_time
```

```python
#Automatic Prediction for time 4,7,6,10,12,13,12,15,12
newdata=pd.Series([4,7,6,10,12,13,12,15,12])
```

```python
data_pred=pd.DataFrame(newdata,columns=['sorting_time'])
```

```python
model.predict(data_pred)
```
