```python
label_encoder = preprocessing.LabelEncoder()
company_files['ShelveLoc']= label_encoder.fit_transform(company_files['ShelveLoc'])
company_files['Urban']= label_encoder.fit_transform(company_files['Urban'])
company_files['US']= label_encoder.fit_transform(company_files['US'])
```

```python
company_files
```

### Splitting dependant and independant variables

```python
d=company_files.iloc[:,1]
x=company_files.iloc[:,1:]
```

```python
x
```

```python
d
```

### Binning the sales data into 4 bins"

```python
y=d.copy()
```

```python
m=pd.cut(y,bins=4)
```

```python
m.value_counts()
```

```python
y
```

```python
y=m
```

### Label encoding for the sales column

```python
y
```

```python
le=preprocessing.LabelEncoder()
```

```python
y=le.fit_transform(y)
```

```python
y
```

### Splitting into train and test data

```python
x_train, x_test,y_train,y_test = train_test_split(x,y, test_size=0.2,random_state=40)
```

### Building model(criteria used is entropy)

```python
model = DecisionTreeClassifier(criterion = 'entropy',max_depth=3)
model.fit(x_train,y_train)
```

### Plotting the decision tree

```python
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (10,10), dpi=600)
tree.plot_tree(model)
```

### The classification is only depended upon the competitor price

```python
x
```

```python
preds = model.predict(x_test) 
pd.Series(preds).value_counts()
```

```python
pd.crosstab(y_test,preds)
```

```python
np.mean(preds==y_test)
```

The classification has a very high accuracy:
